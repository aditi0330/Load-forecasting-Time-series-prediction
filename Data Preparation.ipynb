# Dealing with NAN values

#### dropping 30 june to 7 july..since last rows so doesn't affect the data

```python
indices = data[ (data['year']==2008) & (data['month']==6) & (data['day']==30) ].index
data = data.drop(indices)
```

```python
indices = data[ (data['year']==2008) & (data['month']==7)  ].index
data = data.drop(indices)
```

#### years having NAN values

```python
data[ pd.isnull(data).any(axis=1)].year.value_counts().index
```

#### NAN values should not always imputed by gross mean (e.g.- in case of increasing function)
#### Hence, they can be imputed by mean of surrounding values

#### Imputing NAN values of each zones according to month & year

```python
years=[2005,2006]
for every in years:
    # months having NAN values
    months = data[data['year']==every][ pd.isnull(data).any(axis=1)].month.value_counts().index
    for each in months:
        for i in range(1,21):
            data[(data['year']==every) & (data['month']==each) & (data['zone_id']==i)] = data[(data['year']==every) & (data['month']==each) & (data['zone_id']==i)].fillna(data[(data['year']==every) & (data['zone_id']==i)&(data['month']==each)].mean())
```

```python
data.to_csv("no_nan.csv",index=False)
```

#### Loading temperature file and then merging with original

```python
temperature = pd.read_csv('temperature_history.csv')
```

```python
#dropping rows that are not in load_history
indices = temperature [(temperature['year']==2008) & (temperature['month']==6) & (temperature['day']==30) ].index
temperature = temperature.drop(indices)

indices = temperature[ (temperature['year']==2008) & (temperature['month']==7)  ].index
temperature = temperature.drop(indices)

temperature = temperature.reset_index()
for each in temperature.columns:
    temperature[each] = temperature[each].astype('int32')
    
```

## lets train the model for 5 zones only

```python
df= data [data['zone_id']==1]
for i in range(2,6):
    df= df.append(data [data['zone_id']==i],ignore_index=True)
```

```python
temp= temperature [temperature['station_id']==1]
for i in range(2,6):
    temp= temp.append(temperature [temperature['station_id']==i],ignore_index=True)
```

```python
data = df
temperature = temp
```

```python
for i in range(len(data)): #iterating over all rows
    df=data.iloc[i]     #picking each row of load data
    tem=temperature.iloc[i]     #picking each row of temperature data
    
    for col in col_list:#iterating over each hour
        newdf= pd.DataFrame(index=np.arange(0,1),columns=["zone_id","date_time"])  #new dataframe for each hr
        newdf["zone_id"]=df["zone_id"]
        date=str(df["day"])+'-'+str(df["month"])+'-'+str(df["year"])
        
        newdf["date"]=pd.to_datetime( date)
        newdf["date_time"]=pd.to_datetime( date + " " +  col_time[col])  #date_time of particular hr(col) of particular date(row)
        newdf["usage"]=df[col]     #load of particular hr(col)
        newdf["Temperature"]=tem[col]
        data2=data2.append(newdf,ignore_index=True)    #appending new dataframes to final dataframe
         
```

```python
data2.to_csv('data_for_5_zones.csv',index=False)
```
